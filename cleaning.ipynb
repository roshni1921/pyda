```python
#cleaning
def clean_price_column(price):
    if isinstance(price,str): #check if price is string or not 
        price=price.replace(",","")
        price=price.strip() #remove whitespace
        if price.isnumeric():
            price=int(price) #typecasting
            return price
        if len(price)==0:
            return np.nan
            return price
```

```python
#adding a new column cleaned_price
df["cleaned_price"]=df["price"].apply(clean_price_column)
df
```

```python
df.drop(columns=["price"],inplace=True) #used to remove a particular column
df
```

```python
def clean_rating_column(rating):
    if isinstance(rating,str):
        rating=rating.split()[0]
        return float(rating)
    return rating
```

```python
df["cleaned_rating"]=df["rating"].apply(clean_rating_column)
df
```

```python
df.drop(columns=['rating'],inplace=True)
df
```
