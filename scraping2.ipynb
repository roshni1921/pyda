```python
from dputils.scrape import Tag, Scraper
```

```python
url='https://timesofindia.indiatimes.com/news'
```

```python
news=Scraper(url)
```

```python
target=Tag('ul',cls='HytnJ')
item=Tag("li")
title=Tag('p',cls='CRKrj')
desc=Tag('p',cls='W4Hjm')
```

```python
result=news.get_repeating_page_data(target,item,heading=title,description=desc)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(result)
df
```
